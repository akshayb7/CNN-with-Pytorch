```python
#Import resources
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline

# Read in the image
image = mpimg.imread('images/curved_lane.jpg')

plt.imshow(image)
```

```python
# Convert to grayscale for filtering
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
def filters(img, kernel):
    image = cv2.filter2D(img, -1, kernel)
    plt.imshow(image, cmap='gray')
```

```python
f1 = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

filters(gray, f1)
```

```python
f2 = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])

filters(gray, f2)
```

```python
f = np.array([[0,0,0,-1,1],
             [0,0,-1,2,-1],
              [0,-1,2,-1,0],
              [-1,2,-1,0,0],
             [1,-1,0,0,0]])

filters(gray, f)
```

```python
img2 = mpimg.imread('images/white_lines.jpg')
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(gray2, cmap='gray')
```

```python
filters(gray2, f1)
```

```python
df = np.array([[1.1,2.5,1.1],
             [-1.1,-2.5,-1.1],
             [0,0,0.]])

filters(gray2, df)
```
